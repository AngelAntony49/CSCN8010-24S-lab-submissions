```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<h2 style="text-align: center;">Example 1 </h2>
<p style="text-align: center;">In this i am using Matplotlib and numpy  where i am trying to show sin wave</p>

```python
xpts= np.arange(2,10)
ypts=np.sin(xpts)

plt.plot(xpts,ypts,linestyle='dashed')

plt.title('Fig. 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

<h4> Above exapmle, we gave diagram name as Fig. 1 and we are ploting sin wave through y-axis..</h4>
